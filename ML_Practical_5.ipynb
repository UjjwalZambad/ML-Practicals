```python
transacts = []
for i in range(0, 7501):
    
     transacts.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rule = apriori(transactions = transacts, min_support = 0.003, min_confidence=0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
output=list(rule)
```

```python
output_DataFrame = pd.DataFrame([(tuple(result.items)[0], tuple(result.items)[1],result.support,result.ordered_statistics[0].confidence,result.ordered_statistics[0].lift) for result in output],
                               columns=['Antecedent','Consequent','Support','Confidence','Lift'])`
```
